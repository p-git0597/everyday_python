```python
import requests
import json
```

```python
url = 'https://api.coinbase.com/v2/currencies'
```

```python
response = requests.get(url)
```

```python
currencies = json.loads(response.text)
```

```python
currencies
```

```python
# printing Bitcoin price in different currency
# One can change the currency to anything
exchange_url = 'https://api.coinbase.com/v2/exchange-rates?currency=BTC'
response = requests.get(exchange_url)
exchange_rate = json.loads(response.text)
exchange_rate
```

```python
exchange_rate['data']['rates']['USD']
```

```python
exchange_rate['data']['rates']['INR']
```

```python
# let's get the historical data of coinbase
# :currency_pair = BTC-USD
spot_price_url = 'https://api.coinbase.com/v2/prices/BTC-USD/spot?date=2023-01-01'
response = requests.get(spot_price_url)
spot_price = json.loads(response.text)
spot_price
```
